> **Key terms â€” so this notebook stands on its own** (you shouldn't need the other notebooks to read this one).
>
> - **Effector-uncertain locus** â€” a pigmentation-associated genome location where which gene actually drives the trait is genuinely unknown: the nearest gene is NOT a textbook pigment gene. These 18 loci are the "rescue" targets this diagnostic tests.
> - **Canonical pigmentation gene / effector** (and "canonical neighbor") â€” a gene with a textbook melanin-making role (TYR, OCA2, MC1R, SLC45A2, MITF, TYRP1, SLC24A4â€¦). A locus that connects "only via a canonical neighbor" is explained by the known gene already sitting there, not by anything the screen discovered.
> - **Melanogenesis network / "substrate" / core** â€” the pathway of genes that make melanin, rebuilt in earlier notebooks as a signed (activation/repression), directed graph; here it is the 803-node / 7,819-edge "substrate" graph (from NB7), whose edges are tiered by confidence (T0 Raghunath curation â†’ GRN regulon â†’ OmniPath â†’ STRING). A locus is "rescued" if its candidate gene reaches this core.
> - **Rescue screen & "non-circular route"** â€” the project aim: take a locus whose effector is unknown and try to link it into the melanogenesis network, grading by how many independent evidence lines converge. "Non-circular" means a genuine new link (direct substrate edge, GRN regulon, Reactome subpathway, melanocyte eQTL, â‰¤2-hop path, or verified biology) rather than just re-using the canonical gene already at the locus.
> - **GRN / MITF-SOX10-PAX3 regulon** â€” gene-regulatory network: transcription-factorâ†’target edges, signed activation/repression. MITF, SOX10, and PAX3 are the master pigment transcription factors; their "regulon" is the set of genes they regulate (58 signed edges, built in NB6).
> - **Melanocyte vs bulk-skin eQTL, eGene, coloc (CLPP)** â€” an eQTL is a variant linked to a gene's expression level; an eGene is a gene that has one. Colocalization (coloc) tests whether a GWAS signal and an eQTL share the same causal variant, scored by CLPP (probability; 1.0 = certain shared signal). Using *melanocyte-specific* data (Zhang 2018) instead of bulk skin gives the right cell type, so it can *retract* a wrong causal-gene call or *confirm* a right one.
> - **L2G (locus-to-gene)** â€” Open Targets' score for which gene a GWAS locus most likely acts through; the "provisional bulk L2G resolution" is the pre-melanocyte guess this notebook re-checks.
> - **Convergence grade (A/C)** â€” how much independent evidence a link carries: grade A = several independent lines agree (e.g., LRMDA); grade C = essentially one line. Evidence is only ever ADDED, never used to drop a locus.
> - **STRING / OmniPath** â€” STRING is a database of functional proteinâ€“protein associations (correlation, not mechanism or direction); OmniPath adds curated signed/directed interactions. These are the lowest-confidence graph tiers, so a link resting on STRING alone does not license a mechanistic claim.
> - **Bajpai 2023 CRISPR / casTLE** â€” a genome-wide CRISPR-knockout screen for melanin content (casTLE = its effect-size statistic). A "hit" corroborates a candidate gene; a non-hit is NOT evidence against it (the screen used one cell line).

## Caveats
- **Convergence is thin.** Only LRMDA shows â‰Ą2 independent supporting lines *and* real pigment identity; the rest of the "connected" set rests on a single line (GRN for EGFR, known biology for SIK1/ATRN, weak STRING for CCND1).
- **T3_STRING edges are association, not mechanism.** Every "direct edge to core" via STRING alone (PPP3CAâ†’calmodulin, CCND1â†’CTNNB1) is generic and does not license a melanogenesis claim.
- **GWAS strength and connection strength are decoupled.** ATRN/EGFR connect mechanistically but their Ang2023 signals are null; conversely SYT6/KALRN have significant Ang2023 p-values but no route.
- **Bajpai non-hits are not evidence against** (screen was melanin-content in a specific line).
- Reactome pigmentation gate: 29 genes carry the `Regulation of MITF-M-dependent genes involved in pigmentation` tag; the other 92 are MITF-program-only and are reported separately, never merged.
